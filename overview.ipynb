# Overview

For this task, we used the comments from a [TikTok](https://www.tiktok.com/@kyliedeleonn/video/7568892076381850910?_r=1&_t=ZP-91GtVVLlwvn) displaying the new Starbucks bear cup, which has sparked lots of online conversation. Whereas some are extremely excited to get their hands on this new piece of merchandise (with loyal customers even lining up outside of stores at 3 AM), others are criticising both Starbucks as a company and consumerism as a whole. 

### Research Question
Since both sides of this debate seem extremely passionate about their stance (e.g. "I *HAVE* to get this cup!" vs "Do *NOT* buy this overpriced piece of trash!"), we were wondering:
> Which sides of the debate (pro-bear cup vs. anti-bear cup) exhibit a more aggressive attitude?

### Concepts
Our concepts began as:
- Concept1 = "If the comment isn't negative, code this as 0. If it is, code this as 1."
- Concept2 = "If the comment isn't positive, code this as 0. If it is, code it as 1."

However, after going through some of the comments in the video, we realized that these definitions were not comprehensive enough. For example, what if the comment was expressing desire for the cup (seemingly positive), but also expressing frustration at the limited stock (seemingly negative)? 

Therefore, we adjusted the concepts to be:
| Concept | 1 | 0 |
|----------|----------|----------|
| 1  | Negative comment<br>(mentioning consumerism, <br>criticism against Starbucks, <br>complaints about the limited stock)  | Other<br>(indifference, <br>unrelated to the situation, <br>purely functional questions)  |
| 2  | Positive comment<br>(underlying desire to have the cup, <br>positive emojis, excitement)  | Other<br>(indifference, <br>unrelated to the situation, <br>purely functional questions)  |
