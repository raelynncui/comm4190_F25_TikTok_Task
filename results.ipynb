# Results

After human-coding 100 of the comments and training the LLM to apply that method to 1,000 comments with specific and more accurate prompting, we were able to analyze both the negative (concept 1), positive (concept 2), and even neutral attitudes regarding the debate around the Starbucks Bear cup. From these graphs created by the LLM, we can see that the algorithm coded more of concept 1, so found more negative sentiments, than concept 2, positive sentiments, but not by a significant amount. Neutral comments seemed to dominate the comment section with a coded combination of (0,0), followed by negative comments, then positive comments. It was interesting to see though that the types of comments didn’t seem to vary much more than ~50 comments – which in a dataset of thousands of comments, isn’t much.

We then looked at the interaction type of each comment sentiment; for instance, did positive comments showing support for the Cup receive more likes or replies? What we found was that neutral comments received more likes on average, but neutral and positively associated comments received the same amount of replies. Perhaps people were getting tired of debating in the comments.

Turning to our research question, we originally wondered which type of comment segment expressed more aggressive or strong attitudes and planned to do so by looking at the diction and word choice with each comment. However, as we worked with the data, coding the tonality of certain words seemed to be much more difficult. Thus, we pivoted and instead looked at which types of comments were most common and what type of interactions they were associated with as mentioned before. We had an LLM create another graph, that plots the interactive measures of total likes and replies by concept combination to help visualize this data and draw conclusions – neutral comments were both the most prevalent and received a significantly more number of likes than negative and positive comments, but the overall number of replies no matter the attitude didn’t seem to vary much as represented by the orange bar. These conclusions could tie into many other research questions, for example to question, why is there more variation in like count than reply count on neutral comments? Wouldn’t people want to reply to neutral comments and contribute their opinion? It’s interesting to see what these insights reveal about TikTok users.
